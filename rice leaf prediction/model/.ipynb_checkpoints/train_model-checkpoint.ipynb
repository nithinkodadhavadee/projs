```python
import tensorflow as tf
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras
```

```python
train_dir='/content/drive/MyDrive/deep learning /rice-leaf-diseases-detection-master/rice/train'
test_test='/content/drive/MyDrive/deep learning /rice-leaf-diseases-detection-master/rice/validation'
```

```python
batch_size = 20
epochs = 10
img_height = 224
img_width = 224
```

```python
 train_image_generator = ImageDataGenerator(rescale=1./255)  
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,directory=train_dir,shuffle=True,target_size=(img_height, img_width),class_mode='categorical')
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)  
val_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,directory=test_test,shuffle=True,target_size=(img_height, img_width),class_mode='categorical')
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.inception_v3 import preprocess_input
```

```python
IMAGE_SIZE = [224, 224]
```

```python
base_model = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
cnn_model = keras.models.Sequential([
    base_model,
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(74, activation="relu"),
    keras.layers.Dense(4, activation="softmax")
])
```

```python
cnn_model.compile(optimizer='adam',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
history = cnn_model.fit(train_data_gen, epochs=10,
validation_data= val_data_gen,)
```

```python
cnn_model.save('sandy.h5')
```
